```python
import pandas as pd
data = pd.read_csv("class_marks.csv")

# Define a function to assign grades based on total scores
def assign_grade(total):
    if total >= 30:
        return 'A'
    elif total >= 25:
        return 'B'
    elif total >= 20:
        return 'C'
    elif total >= 15:
        return 'D'
    else:
        return 'F'

data['Grade'] = data['Total'].apply(assign_grade)
output_file_path = 'class_marks.csv'
data.to_csv(output_file_path, index=False)

data
```
